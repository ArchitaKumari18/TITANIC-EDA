```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Make charts look nice
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("../data/titanic.csv")

df.head()
```

```python
df = pd.read_csv("../data/titanic.csv")
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x="Survived",
    data=df,
    hue="Survived",
    palette="Set2",
    legend=False
)

plt.title("Number of Survivors vs Non-Survivors", fontsize=16, fontweight="bold")
plt.xlabel("Survival Status", fontsize=12)
plt.ylabel("Number of Passengers", fontsize=12)

plt.xticks([0, 1], ["Did Not Survive", "Survived"])

plt.tight_layout()
plt.show()
```

### Takeaway
More passengers did not survive than survived, highlighting the severity of the Titanic disaster.

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x="Sex",
    hue="Survived",
    data=df,
    palette="Set1"
)

plt.title("Survival by Gender", fontsize=16)
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")

plt.legend(title="Survived", labels=["No", "Yes"])

plt.show()
```

### Takeaway

Female passengers had a much higher survival rate than male passengers.

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x="Pclass",
    hue="Survived",
    data=df,
    palette="viridis"
)

plt.title("Survival by Passenger Class", fontsize=16)
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")

plt.show()
```

### Takeaway

Passengers in first class had the highest survival rate.

```python
# Fill missing Age values
df["Age"] = df["Age"].fillna(df["Age"].median())

plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x="Age",
    bins=30,
    kde=True,
    color="skyblue"
)

plt.title("Age Distribution of Titanic Passengers", fontsize=16)
plt.xlabel("Age")
plt.ylabel("Number of Passengers")

plt.show()
```

### Takeaway

Most passengers were between 20 and 40 years old.

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x="Fare",
    bins=30,
    color="green"
)

plt.title("Ticket Fare Distribution", fontsize=16)
plt.xlabel("Fare")
plt.ylabel("Number of Passengers")

plt.show()
```

### Takeaway

Most passengers paid lower ticket fares, while only a few paid very high fares.

```python
plt.figure(figsize=(8,6))

# Select only numeric columns
numeric_df = df.select_dtypes(include="number")

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap of Titanic Dataset", fontsize=16)

plt.show()
```

### Takeaway

Passenger class, fare, and survival show meaningful relationships in the dataset.

# Titanic Survival Story

This analysis explores the survival patterns of passengers aboard the Titanic using six different visualizations. The first chart shows that more passengers died than survived, highlighting the seriousness of the disaster. The second chart reveals that women had a much higher survival rate than men, showing that gender played an important role during rescue efforts. The third chart demonstrates that passengers traveling in first class had better survival chances than those in second or third class.

The age distribution shows that most passengers were young adults between 20 and 40 years old. The fare distribution indicates that the majority of passengers purchased lower-priced tickets, while only a small number paid very high fares. Finally, the correlation heatmap helps identify relationships among numerical variables, showing that passenger class and fare were related to survival.

Overall, the analysis suggests that survival depended on several important factors, including gender, passenger class, and ticket fare. These six publication-quality charts clearly communicate the story hidden in the Titanic dataset. By presenting clear titles, labeled axes, and short explanations, the notebook effectively transforms raw data into meaningful insights and demonstrates the importance of data visualization in exploratory data analysis.
